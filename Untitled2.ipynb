```python
import cv2
import matplotlib.pyplot as plt
import skimage.io as imageio
import numpy as np
import pandas as pd
```

```python
mar=cv2.imread("C:/Users/se7en/Desktop/dd/img_data/undrowsiness/israil_undrowsiness_1460.jpg")
imageio.imshow(mar)
```

```python
rgb=cv2.cvtColor(mar,cv2.COLOR_BGR2RGB)
imageio.imshow(rgb)
```

```python
eyedetect=cv2.CascadeClassifier('data/haarcascade_eye.xml')
geteye=eyedetect.detectMultiScale(rgb)
geteye
gozler=[]
for (x,y,w,h) in geteye:
    gozler.append(rgb[y:y+h, x:x+w])
imageio.imshow(gozler[0])
```
